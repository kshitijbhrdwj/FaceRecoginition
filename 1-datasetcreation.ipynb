```python
import os

data_path = "images"
labels = os.listdir(data_path)
labels
```

```python
categories = [i for i in range(len(labels))]
category_dict = dict(zip(labels,categories))
category_dict
```

```python
# after assigning categories we move to process the image

import cv2

data= []
target = []

# loading haarcascade classifier
face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
for label in labels:
    imgs_path = os.path.join(data_path,label)
    img_names = os.listdir(imgs_path)             # img_names consist of image names for each folders(labels)
    
    for img_name in img_names:
        img_path = os.path.join(imgs_path,img_name)  # path for each image
        img = cv2.imread(img_path)    
        
        #cv2.imshow('LIVE',img)                   # we can see each image getting loaded
        #key = cv2.waitKey(100)
        #if key == 27:
        #     break
#cv2.destroyAllWinsdows()
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#         cv2.imshow('LIVE',img)
#         cv2.imshow('GRAY',gray)
#         key = cv2.waitKey(200)
#         if key == 27:
#              break
        faces = face_classifier.detectMultiScale(gray)
        for (x,y,w,h) in faces:
#             cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2) 
#             cv2.imshow('LIVE',img)
#             key = cv2.waitKey(500)
#             if key == 27:
#                  break
        # now we crop the section detected by our classifier as face
            cropped_face = gray[y:y+h,x:x+w]
        
        # now we resize all img to commmon dimesnsion
            cropped_face = cv2.resize(cropped_face,(100,100))
            cv2.imshow('LIVE',cropped_face)
            key = cv2.waitKey(500)
            if key == 27:
                 break
            ip = int(input('Valid Image? [Yes:1, No:0]'))
            if ip == 1:
                data.append(cropped_face)
                target.append(category_dict[label])
            else: 
                continue
```

```python
# get image path
for label in labels:
    imgs_path = os.path.join(data_path,label)
    img_names = os.listdir(imgs_path)             # img_names consist of image names for each folders(labels)
    
    for img_name in img_names:
        img_path = os.path.join(imgs_path,img_name)  # path for each image
        img = cv2.imread(img_path)    
        
#         cv2.imshow('LIVE',img)                   # we can see each image getting loaded
#         key = cv2.waitkey(100)
#         if key == 27:
#             break
#cv2.destroyAllWinsdows()

# after loading each image we apply preprocessing. Firstly, we convert these RGB image to Gray
        
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        
#         cv2.imshow('LIVE',img)
#         cv2.imshow('GRAY',gray)
#         key = cv2.waitKey(100)
#         if key == 27:
#             break

# Now we load the harcascade classifier which is present in xml to us. After loading we use the face classifier to our gray images

        faces = face_classifier.detectMultiScale(gray)
    
        
        for (x,y,w,h) in faces:
            
            # sometimes it detects other objcts as faces, we draw a rectangle to test if its detecting faces correctly
            
               # coloured_img, x1,y1,  x2,y2,  green color,  pizel width 
#             cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)    
#             cv2.imshow('LIVE',img)
#             key = cv2.waitKey(100)
#             if key == 27:
#                 break

            # now we crop the section detected by our classifier as face
            cropped_face = gray[y:y+h,x:x+w]
        
            # now we resize all img to commmon dimesnsion
            cropped_face = cv2.resize(cropped_face,(50,50))
            
            # we can ask now the user to classify if the cropped imag should be taken as dataset or not and based on that 
            # append the img to data list and label to target list
            data.append(cropped_face)
            target.append(category_dict[label])
```

```python
# after having data and target lists, we flatten them since they consist of images in 2d form i.e. for each img - 50*50
import numpy as np
data = np.array(data)
target = np.array(target)

no_of_images,height,width = data.shape
print(data.shape)
print(target.shape)
data = data.reshape(no_of_images,height/2*width/2)   # for eg - > 400, 50,50 to 400, 2500

np.save('data',data)
np.save('target',target)


# now at last we save te data set as np array. 
# data consist of flatten image prompted from the user whther those should be included in th data set or not.
# that data is already present... but this is how overall we create our dataset.
```
